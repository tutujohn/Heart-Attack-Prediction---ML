```python
#Import relevant libraries and modules
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error

import joblib
```

# Data Preproscesing

```python
#Drop Patient ID and Heart Attack
y = data["Heart Attack Risk"]
X = data.drop(["Patient ID", "Heart Attack Risk"], axis = 1)
```

```python
#Remove Blood Pressure
Categorical_Column = X.select_dtypes(include = object).columns
Categorical_Column = [x for x in Categorical_Column if x != "Blood Pressure"]
Categorical_Column
```

```python
#One hot encoding categorical value

X = pd.get_dummies(X, prefix=Categorical_Column, columns=Categorical_Column)
```

```python
X["Blood Pressure"].head()
```

```python
#Slit and convect Blood Prssure to float
def process_rec(col):
  res = []
  for rec in col:
    m = [int(x) for x in rec.split("/")]
    if len(m) == 2:
      m = m[0] / m[1]
    else:
      m = m[0]

    res.append(m)
  return res

X["Blood Pressure"] = process_rec(X["Blood Pressure"])
```

```python
X.head()
```

## Modelling

```python
#Slipt data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
#Impliment Random Forest Classifier
model = RandomForestClassifier(random_state=100)
model.fit(X_train, y_train)

preds = model.predict(X_test)
```

```python
#Evaluate the perfromance
mse = mean_squared_error(y_test, preds)
mse
```
